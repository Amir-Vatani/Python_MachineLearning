```python
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
```

```python
np.random.seed(121)

N = 500
std = 0.6
melons = pd.DataFrame({
    'height': np.random.normal(8, std, N),
    'width': np.random.normal(5, std, N),
    'weight': np.random.normal(10, std, N)}
)

ballons = pd.DataFrame({
    'height': np.random.normal(9, std, N),
    'width': np.random.normal(7, std, N),
    'weight': np.random.normal(1, std, N)}
)

mel_bal = pd.concat([melons, ballons])
```

```python
fig = plt.figure()
threeD = fig.add_subplot(projection='3d')

threeD.scatter3D(melons['height'],melons['width'],melons['weight'], color = "orange")
threeD.scatter3D(ballons['height'],ballons['width'],ballons['weight'], color = "blue")

threeD.set_xlabel('height')
threeD.set_ylabel('width')
threeD.set_zlabel('weight')
plt.show()
```
