```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from scipy import stats
from numpy.linalg import inv
```

```python
X = np.random.uniform(0, 75, 200) # study hour 
Y = X + np.random.uniform(0, 25, 200) # grade
plt.scatter(X ,Y)
plt.show()
```

```python
Y_pred = X * 0.8

plt.scatter(X ,Y)
plt.plot(X , Y_pred, c = 'red', lw = 3)
plt.show()
```

```python
x = X.reshape((200,1))
y = Y.reshape((200,1))
m = np.matmul(inv(np.matmul(x.T, x)), np.matmul(x.T, y))
m = m[0,0]
m
```

```python
Y_pred = m * X
```

```python
plt.scatter(X ,Y)
plt.plot(X , Y_pred, c = 'red', lw = 3)
plt.show()
```

```python
linregress = stats.linregress(X, Y)
```

```python
plt.scatter(X ,Y, label='orginal data')
plt.plot(X , Y_pred, c = 'red', lw = 3, label='linear least squares line')
plt.plot(X , linregress.intercept + linregress.slope * X, c = 'yellow', lw = 3, label='linregress line')
plt.legend(loc='best')
plt.show()
```
